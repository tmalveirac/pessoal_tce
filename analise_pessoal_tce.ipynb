```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pessoal_df = pd.read_excel("./dados/ServidoresEfetivosEComissionados.xls")
```

```python
pessoal_df
```

```python
qtd_cargos = pessoal_df.groupby('CARGO').CARGO.count().sort_values(ascending=False) #agg(['count']).sort_values(by='count', ascending=False)
```

```python
qtd_cargos
```

```python
qtd_cargos.plot(kind='bar', grid=True)
```

```python
is_analista = pessoal_df['CARGO'] == 'Analista de Controle Externo'
is_tecnico = pessoal_df['CARGO'] == 'Técnico de Controle Externo'
```

```python
qtd_classe_analista = pessoal_df[is_analista].groupby('REFERENCIA_CLASSE').REFERENCIA_CLASSE.count().sort_values(ascending=False)
#qtd_classe_analista.dropna(inplace=True)
```

```python
qtd_classe_analista
```

```python
qtd_classe_analista.plot(kind='bar', grid=True, title='Quantidade de Analistas por REF TCE-CE')
```

```python
qtd_classe_tecnicos = pessoal_df[is_tecnico].groupby('REFERENCIA_CLASSE').REFERENCIA_CLASSE.count().sort_values(ascending=False)
```

```python
qtd_classe_tecnicos
```

```python
qtd_classe_tecnicos.plot(kind='bar', grid=True, title='Quantidade de Técnicos por REF TCE-CE')
```
